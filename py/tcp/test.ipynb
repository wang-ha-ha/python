```python
def binary_search(list,item):
    low = 0
    high = len(list) - 1

    while low <= high:
        mid = (low + high) // 2
        guess = list[mid]
        if guess == item:
            return mid
        elif guess > item:
            high = mid - 1
        elif guess < item:
            low = mid + 1
    
    return None

my_list = [1, 3, 5, 7, 9]

print(binary_search(my_list,3))
print(binary_search(my_list,-1))
```

```python
def sum(arr):
    if (arr == []):
        return 0
    else:
        return arr[0] + sum(arr[1:])

def count(arr):
    if (arr == []):
        return 0
    return 1 + count(arr[1:])

def max(arr,max_value):
    if (arr == []):
        return max_value
    
    if (max_value < arr[0]):
        return max(arr[1:],arr[0])
    else:
        return max(arr[1:],max_value)

my_arr = [1,2,3,4,5,6,7,8,9]

print(sum(my_arr))
print(count(my_arr))
print(max(my_arr,my_arr[0]))
```

```python
import random

def qsort(arr):
    if len(arr) < 2:
        return arr

    n = []
    l = []

    for i in range(1, len(arr)):
        if arr[i] <= arr[0]:
            n.append(arr[i])
        elif arr[i] > arr[0]:
            l.append(arr[i])
    
    return qsort(n) + [arr[0]] + qsort(l)

def gen_test_data(arr,len):
    for i in range(len):
        arr.append(random.randrange(0, 1000))

my_arr = []
gen_test_data(my_arr,100)
# my_arr.clear()
print(my_arr.sort())
print(my_arr)
my_arr = qsort(my_arr)
print(my_arr)
```

```python
import sys

def coin_change(coins,amount):
    _min = sys.maxsize

    def dfs(_count,_sum):        
        nonlocal _min
        
        if(_sum == amount):
            if(_count < _min):
                _min = _count
            return
        elif(_sum > amount):
            return
        
        for i in range(len(coins)):
            dfs(_count + 1,coins[i] + _sum)

    def dp(_sum):
        res = -1
        
        for i in coins:
            res = min(res,1 + dp(_sum - i))
        
        return res
    # dfs(0,0)
    # return _min

    return dp(amount)

coins = [1,5,10,100]
r = coin_change(coins,21231233)
print(r)
```
